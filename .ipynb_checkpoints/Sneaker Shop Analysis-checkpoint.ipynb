<p><b> **On Shopify, we have exactly 100 sneaker shops, and each of these shops sells only one model of shoe. We want to do some analysis of the average order value (AOV). When we look at orders data over a 30 day window, we naively calculate an AOV of $3145.13. Given that we know these shops are selling sneakers, a relatively affordable item, something seems wrong with our analysis.** </b></p>

 - <b> Think about what could be going wrong with our calculation. Think about a better way to evaluate this data. </b>
  - The unit price of a sneaker in Shop 78 is 25,725 USD which has increased the AOV. Moreover, shop 42 has an order value of 704000 for 2000 shoes. These act as an outlier. In order to get a proper estimate, these shops can be removed off the list.
  
 - <b> What metric would you report for this dataset? </b>
  - Mean is non-trivially influenced by an outlier. Therefore, using <b>Median</b> as a metric would suit this dataset.
 - <b> What is its value? </b>
  - The Median Order Value is <b> 284 USD </b>
